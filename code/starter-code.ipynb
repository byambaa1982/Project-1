```python
def find_std(any_list):
    mean=np.mean(any_list)
    st=np.sqrt(sum([(i-mean)**2 for i in any_list])*1/len(any_list))
    return st
find_std(my_sat['Math'])
    
```

##### 26. Execute a hypothesis test comparing the SAT and ACT participation rates. Use $\alpha = 0.05$. Be sure to interpret your results.

```python
# Initialize a matplotlib "figure:"
fig = plt.figure(figsize=(10,5))
ax = fig.gca()

zx = np.linspace(np.mean(rate_act)-3.5*np.std(rate_sat)/np.sqrt(len(rate_sat)), np.mean(rate_act)+3.5*np.std(rate_sat)/np.sqrt(len(rate_sat)), 250)

Z = stats.norm(np.mean(rate_act), np.std(rate_sat)/np.sqrt(len(rate_sat)))
zy = Z.pdf(zx)

# 90%:
ax.axvline(np.mean(rate_act)-1.645*np.std(rate_sat)/np.sqrt(len(rate_sat)), ls='dashed', lw=3, color='#333333', alpha=0.7)
ax.axvline(np.mean(rate_act)+1.645*np.std(rate_sat)/np.sqrt(len(rate_sat)), ls='dashed', lw=3, color='#333333', alpha=0.7, label='z=1.645, 90%')

# 95%:
ax.axvline(np.mean(rate_act)-1.96*np.std(rate_sat)/np.sqrt(len(rate_sat)), ls='dashed', lw=3, color='#666666', alpha=0.7)
ax.axvline(np.mean(rate_act)+1.96*np.std(rate_sat)/np.sqrt(len(rate_sat)), ls='dashed', lw=3, color='#666666', alpha=0.7, label='z=1.96, 95%')

# 99%:
ax.axvline(np.mean(rate_act)-2.575*np.std(rate_sat)/np.sqrt(len(rate_sat)), ls='dashed', lw=3, color='#999999', alpha=0.7)
ax.axvline(np.mean(rate_act)+2.575*np.std(rate_sat)/np.sqrt(len(rate_sat)), ls='dashed', lw=3, color='#999999', alpha=0.7, label='z=2.575, 99%')

ax.plot(zx, zy, linewidth=3, color='darkred')
ax.legend(loc='upper left')
```
